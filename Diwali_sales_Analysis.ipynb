#### DIWALI SALES ANALYSIS


```python
# importing libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df= pd.read_csv('Diwali Sales Data.csv', encoding ='unicode escape')
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#rename_column
df.rename(columns={'Marital_status':'Shaadi'})
```

```python
df.info()
```

```python
#dropping null columns
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
#checking null values
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
pd.isnull(df).sum().sum()
```

```python
df.dropna(inplace=True)
```

#### Exploratory Data Analysis

```python
## plotting a bar chart for gender and its chart
ax = sns.countplot(x='Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting a barchar for gender and sales
sales_gen = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

##### most buyers are females than males

```python
#age
ax = sns.countplot(x ='Age Group', data = df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

#Females between age 26-35 are the one who buys most whereas the females between age 0-17 are the ones who buys less.

### Age group vs sales


```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending =False)
sns.barplot(x ='Age Group', y ='Amount', data = sales_age )
```

### From this chart it is clear that most buyers are from age 26-35

### State

```python
#TOTAL NUMBER OF ORDERS FROM TOP 10 STATES
state_sales = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by="Orders", ascending =False)
sns.set_theme(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State',  y = 'Orders', data =state_sales )
```

#### From the above chart we can see that most of the buyers are frim uttar pradesh

### Marital Status

```python
df.head()
```

```python
ax= sns.countplot(data =df,  x ='Marital_Status')
sns.set_theme(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
### Marital status of male and female
gender_mar= df.groupby(['Marital_Status'], as_index = False)['Gender'].sum().sort_values(by = 'Gender', ascending = False)
sns.set_theme(rc={'figure.figsize':(6,5)})

sns.barplot(x ='Marital_Status', y = 'Gender', data = gender_mar)
```

```python
#sales on the basis of marital status of male and female
sales_mar = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)

sns.set_theme(rc={'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status', y ='Amount', hue = 'Gender', data = sales_mar)
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*


#### occupation

```python
ax = sns.countplot(data = df, x ='Occupation')
sns.set_theme(rc={'figure.figsize':(1,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#sales on the basis of occupation
sales_occupation= df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending =False)
sns.set_theme(rc ={'figure.figsize':(20,5)})
sns.barplot(x ='Occupation', y ='Amount', data = sales_occupation)
```

### Most buyers are from It sector

#### product category

```python
sns.set_theme(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data= df, x ='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
### prodcut sale
sales_product = df.groupby(['Product_Category'], as_index =False)['Amount'].sum().sort_values(by ='Amount', ascending = False)
sns.set_theme(rc={'figure.figsize':(20,5)})
sns.barplot(x ='Product_Category', y = 'Amount', data = sales_product)
```

### This graph shows that people buy food items the most
